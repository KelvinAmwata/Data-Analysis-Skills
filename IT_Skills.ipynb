```python
from matplotlib import colors
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
#Upload data to your google drive
df=pd.read_csv("/content/drive/MyDrive/ITS - HelpDesk.csv")
print(df)
# set the column as index to use as plot labels
df.index=df['Skill']
#df.index = df['Skill']
#sorting by descending order
df=df.sort_values(by='Frequency',ascending=False)
#Use cumsum function to create cumulative percentage column
df["cumulativePercentage"]=df["Frequency"].cumsum()/df["Frequency"].sum()*100
#Ensure only two decimal points are shown
df["cumulativePercentage"]=df["cumulativePercentage"].apply(lambda x: round(x,2))
print(df)
fig,ax=plt.subplots()
ax.bar(df.index,df["Frequency"],color="C0")
#Show labels diagonally 
ax.set_xticklabels(df['Skill'],rotation=45)
#ax2 and ax1 will be shown at the same time with twinx
ax2=ax.twinx()

ax2.plot(df.index,df["cumulativePercentage"],color="C1",marker="D",ms=7)

ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")

ax2.tick_params(axis="y", colors="C1")

for index, v in df.iterrows():
  label = round(v['cumulativePercentage'], 1)
  plt.annotate(label, xy=(v['Skill'],v['cumulativePercentage']))

plt.show()

```
